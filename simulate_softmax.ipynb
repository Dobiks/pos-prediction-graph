```python
import pandas as pd
import numpy as np
import random
```

```python
path = "csv_files/detection_graph.csv"
```

```python
df = pd.read_csv(path)
```

```python
df
```

```python
#for each element in 'name' column create a column with the name
for i in range(len(df)):
    df[df['name'][i]] = 0
    
```

```python
names_list = df['name'].tolist()
```

```python
tmp = [1,2,3,4]
print(f"len(tmp): {len(tmp)}")
for i,v in enumerate(tmp):
    print(i,v)
```

```python
# for row
for i in range(len(df)):
    name = df["name"][i]
    current_softmax = random.uniform(0.51, 0.9)
    df[name][i] = current_softmax
    random_names = random.sample(names_list, random.randint(1, len(names_list)))
    if name in random_names:
        random_names.remove(name)
    other_values = [random.randint(1,100) for i in range(len(random_names))]
    #normalize to 1
    other_values = [(float(i)/np.sum(other_values))*(1-current_softmax) for i in other_values]
    for j in range(len(random_names)):
        df[random_names[j]][i] = other_values[j]
```

```python
df
```

```python
#check if sum of each row is 1
for i in range(len(df)):
    #sum colums value from names_lsit
    sum = 0
    for name in names_list:
        sum += df[name][i]
    print(f"sum: {sum} for row: {i}")
```

```python
df.to_csv("csv_files/detection_graph_softmax.csv", index=False)
```
