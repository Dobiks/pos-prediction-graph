```python
import pandas as pd
import numpy as np
import random
```

```python
final_df = pd.DataFrame()
```

```python
def add_softmax():
    path = "csv_files/randomized_graph.csv"
    df = pd.read_csv(path)

    for i in range(len(df)):
        df[df["name"][i]] = 0

    names_list = df["name"].tolist()
    # for row
    for i in range(len(df)):
        name = df["name"][i]
        current_softmax = random.uniform(0.8, 0.99)
        df[name][i] = current_softmax
        random_names = random.sample(names_list, random.randint(1, len(names_list)))
        if name in random_names:
            random_names.remove(name)
        other_values = [random.randint(1, 100) for i in range(len(random_names))]
        other_values = [
            (float(i) / np.sum(other_values)) * (1 - current_softmax)
            for i in other_values
        ]
        for j in range(len(random_names)):
            df[random_names[j]][i] = other_values[j]

    # for i in range(len(df)):
    #     #sum colums value from names_lsit
    #     sum = 0
    #     for name in names_list:
    #         sum += df[name][i]
    #     print(f"sum: {sum} for row: {i}")

    return df
```

```python
import random
from randomize import randomize
# n = random.randint(3, 8)
n = 10
for i in range(n):
    randomize()
    df = add_softmax()
    final_df = final_df.append(df)
#reset index
final_df = final_df.reset_index(drop=True)
```

```python
#drop name column
final_df = final_df.drop(columns=["name"])
#move index to first column "id"
final_df.insert(0, "id", final_df.index)
final_df
```
