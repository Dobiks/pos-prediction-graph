```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import re
```

```python
def extract_number(filename):
    match = re.search(r'results_(\d+).csv', filename)
    if match:
        return int(match.group(1))
    return None
```

```python
result_dir = 'results/08'
files = [f for f in os.listdir(result_dir) if f.startswith('results')]
files
```

```python
dfs = {}
for file in files:
    df = pd.read_csv(os.path.join(result_dir, file))
    dfs[extract_number(file)] = df
    
```

```python
#plot first
fig, ax = plt.subplots(figsize=(10, 5))
for key, df in dfs.items():
    ax.plot(df['distances'], df['component_mae'], label=key)
ax.set_xlabel('Distance')
ax.set_ylabel('MAE')
ax.legend()
plt.show()

```
