```python
import networkx as nx
from graph_utils import plot_graph
import numpy as np
import pandas as pd
```

```python
static = pd.read_csv('static_graph.csv')
static['name'] = static['name'].str.replace(' ', '')
```

```python
# add column 'distance_to_parent'
# static["t_x"] = 0
# static["t_y"] = 0
# static["t_z"] = 0
static["distance_to_parent"] = 0
for i in range(len(static)):
    if static["parent"][i] != 0 and static["parent"][i] != "0":
        parent_xyz = [
            static[static["name"] == static["parent"][i]]["x"].values[0],
            static[static["name"] == static["parent"][i]]["y"].values[0],
            static[static["name"] == static["parent"][i]]["z"].values[0],
        ]
        node_xyz = [
            static[static["name"] == static["name"][i]]["x"].values[0],
            static[static["name"] == static["name"][i]]["y"].values[0],
            static[static["name"] == static["name"][i]]["z"].values[0],
        ]
        node_name = static["name"][i]
        # transformation = np.array(node_xyz) - np.array(parent_xyz)
        # # print(transformation)
        # static["t_x"][i] = transformation[0]
        # static["t_y"][i] = transformation[1]
        # static["t_z"][i] = transformation[2]

        dist_to_parent = np.sqrt(
            (node_xyz[0] - parent_xyz[0]) ** 2
            + (node_xyz[1] - parent_xyz[1]) ** 2
            + (node_xyz[2] - parent_xyz[2]) ** 2
        )
        #round to roof
        dist_to_parent = np.round(dist_to_parent, 0)
        static["distance_to_parent"][i] = dist_to_parent

        
```
